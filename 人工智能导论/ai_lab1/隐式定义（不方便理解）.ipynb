````python
# -*- coding: utf-8 -*-
"""
çŒ´å­æ‘˜é¦™è•‰é—®é¢˜â€”â€”ä½¿ç”¨äº§ç”Ÿå¼ç³»ç»Ÿæ€æƒ³æ±‚è§£

é—®é¢˜æè¿°ï¼š
  ä¸€ä¸ªæˆ¿é—´é‡Œå¤©èŠ±æ¿ä¸ŠæŒ‚ç€é¦™è•‰ï¼Œæˆ¿é—´å†…æœ‰ä¸€åªçŒ´å­å’Œä¸€ä¸ªå¯ç§»åŠ¨çš„ç®±å­ã€‚
  çŒ´å­å¿…é¡»æ¨ç®±å­åˆ°é¦™è•‰ä¸‹æ–¹ï¼Œå†æ”€ä¸Šç®±å­æ‰èƒ½æ‘˜åˆ°é¦™è•‰ã€‚
  
åˆå§‹çŠ¶æ€ï¼š
  çŒ´å­ä½ç½®ä¸º 'A'
  ç®±å­ä½ç½®ä¸º 'C'
  é¦™è•‰ä½ç½®å›ºå®šåœ¨ 'B'
  çŒ´å­ä¸åœ¨ç®±å­ä¸Šï¼Œä¸”å°šæœªæ‘˜åˆ°é¦™è•‰

ç›®æ ‡çŠ¶æ€ï¼š
  çŒ´å­æ‘˜åˆ°é¦™è•‰
"""

class State:
    def __init__(self, monkey, box, on_box, has_banana):
        self.monkey = monkey      # çŒ´å­ä½ç½®
        self.box = box            # ç®±å­ä½ç½®
        self.on_box = on_box      # çŒ´å­æ˜¯å¦ç«™åœ¨ç®±å­ä¸Šï¼ˆTrue/Falseï¼‰
        self.has_banana = has_banana  # æ˜¯å¦å·²æ‘˜åˆ°é¦™è•‰ï¼ˆTrue/Falseï¼‰
    
    def __eq__(self, other):
        return (self.monkey == other.monkey and 
                self.box == other.box and
                self.on_box == other.on_box and 
                self.has_banana == other.has_banana)
    
    def __hash__(self):
        return hash((self.monkey, self.box, self.on_box, self.has_banana))
    
    def __repr__(self):
        return (f"State(monkey={self.monkey}, box={self.box}, "
                f"on_box={self.on_box}, has_banana={self.has_banana})")

def actions(state):
    """
    æ ¹æ®å½“å‰çŠ¶æ€è¿”å›å¯é‡‡å–çš„åŠ¨ä½œåˆ—è¡¨ã€‚
    å®šä¹‰çš„åŠ¨ä½œåŠæ¡ä»¶ï¼š
      1. Move: çŒ´å­ä¸åœ¨ç®±å­ä¸Šæ—¶ï¼Œå¯ç§»åŠ¨åˆ°ä»»ä¸€ä¸åŒä½ç½®ã€‚
      2. Push: å½“çŒ´å­ä¸ç®±å­åœ¨åŒä¸€ä½ç½®ä¸”ä¸åœ¨ç®±å­ä¸Šæ—¶ï¼Œå¯æ¨åŠ¨ç®±å­åˆ°å…¶å®ƒä½ç½®ï¼ŒåŒæ—¶çŒ´å­è·Ÿéšç®±å­ç§»åŠ¨ã€‚
      3. ClimbUp: å½“çŒ´å­ä¸ç®±å­åœ¨åŒä¸€ä½ç½®ä¸”ä¸åœ¨ç®±å­ä¸Šæ—¶ï¼Œå¯çˆ¬ä¸Šç®±å­ã€‚
      4. ClimbDown: å½“çŒ´å­åœ¨ç®±å­ä¸Šæ—¶ï¼Œå¯ä¸‹ç®±å­ï¼ˆå›åˆ°ç®±å­åŒä¸€ä½ç½®ï¼‰ã€‚
      5. Grasp: å½“çŒ´å­åœ¨ç®±å­ä¸Šï¼Œä¸”çŒ´å­ä½ç½®æ­£å¥½åœ¨é¦™è•‰æ‰€åœ¨ä½ç½®ï¼ˆå›ºå®šä¸º 'B'ï¼‰æ—¶ï¼Œå¯æ‘˜é¦™è•‰ã€‚
    """
    available_actions = []
    locations = ['A', 'B', 'C']
    # åŠ¨ä½œï¼šMoveï¼ˆç§»åŠ¨ï¼‰ï¼Œæ¡ä»¶ï¼šçŒ´å­ä¸åœ¨ç®±å­ä¸Š
    if not state.on_box:
        for loc in locations:
            if loc != state.monkey:
                available_actions.append(("Move", loc))
    # åŠ¨ä½œï¼šPushï¼ˆæ¨ç®±å­ï¼‰ï¼Œæ¡ä»¶ï¼šçŒ´å­å’Œç®±å­åœ¨åŒä¸€ä½ç½®ä¸”çŒ´å­ä¸åœ¨ç®±å­ä¸Š
    if not state.on_box and state.monkey == state.box:
        for loc in locations:
            if loc != state.box:
                available_actions.append(("Push", loc))
    # åŠ¨ä½œï¼šClimbUpï¼ˆçˆ¬ä¸Šç®±å­ï¼‰ï¼Œæ¡ä»¶ï¼šçŒ´å­å’Œç®±å­åœ¨åŒä¸€ä½ç½®ä¸”ä¸åœ¨ç®±å­ä¸Š
    if not state.on_box and state.monkey == state.box:
        available_actions.append(("ClimbUp", None))
    # åŠ¨ä½œï¼šClimbDownï¼ˆä¸‹ç®±å­ï¼‰ï¼Œæ¡ä»¶ï¼šçŒ´å­åœ¨ç®±å­ä¸Š
    if state.on_box:
        available_actions.append(("ClimbDown", None))
    # åŠ¨ä½œï¼šGraspï¼ˆæ‘˜é¦™è•‰ï¼‰ï¼Œæ¡ä»¶ï¼šçŒ´å­åœ¨ç®±å­ä¸Šä¸”çŒ´å­ä½ç½®ç­‰äºé¦™è•‰æ‰€åœ¨ä½ç½®ï¼ˆ'B'ï¼‰
    if state.on_box and state.monkey == 'B' and not state.has_banana:
        available_actions.append(("Grasp", None))
    return available_actions

def result(state, action):
    """
    æ ¹æ®åŠ¨ä½œäº§ç”Ÿæ–°çš„çŠ¶æ€ã€‚åŠ¨ä½œæ˜¯ä¸€ä¸ªäºŒå…ƒç»„ (åŠ¨ä½œåç§°, å‚æ•°)ã€‚
    """
    act, param = action
    # å¤åˆ¶å½“å‰çŠ¶æ€ï¼Œé¿å…ä¿®æ”¹åŸçŠ¶æ€
    new_state = State(state.monkey, state.box, state.on_box, state.has_banana)
    
    if act == "Move":
        new_state.monkey = param
    elif act == "Push":
        # æ¨åŠ¨ç®±å­ï¼šçŒ´å­ä¸ç®±å­åŒæ—¶ç§»åŠ¨åˆ°ç›®æ ‡ä½ç½®
        new_state.box = param
        new_state.monkey = param
    elif act == "ClimbUp":
        new_state.on_box = True
    elif act == "ClimbDown":
        new_state.on_box = False
    elif act == "Grasp":
        new_state.has_banana = True
    return new_state

def is_goal(state):
    """åˆ¤æ–­æ˜¯å¦è¾¾åˆ°ç›®æ ‡çŠ¶æ€ï¼šæ‘˜åˆ°é¦™è•‰ã€‚"""
    return state.has_banana

def bfs(initial_state):
    """
    ä½¿ç”¨å¹¿åº¦ä¼˜å…ˆæœç´¢å¯»æ‰¾ä¸€æ¡åŠ¨ä½œåºåˆ—ä½¿çŒ´å­æ‘˜åˆ°é¦™è•‰ã€‚
    è¿”å›åŠ¨ä½œåºåˆ—ï¼ˆåˆ—è¡¨ï¼‰ï¼Œè‹¥æ— è§£åˆ™è¿”å› Noneã€‚
    """
    from collections import deque
    frontier = deque()
    frontier.append((initial_state, []))
    explored = set()
    
    while frontier:
        state, path = frontier.popleft()
        if is_goal(state):
            return path
        explored.add(state)
        for act in actions(state):
            new_state = result(state, act)
            if new_state not in explored:
                frontier.append((new_state, path + [act]))
    return None

if __name__ == '__main__':
    # åˆå§‹çŠ¶æ€ï¼šçŒ´å­åœ¨ 'A'ï¼Œç®±å­åœ¨ 'C'ï¼ŒçŒ´å­ä¸åœ¨ç®±å­ä¸Šï¼Œä¸”å°šæœªæ‘˜åˆ°é¦™è•‰ï¼ˆé¦™è•‰å›ºå®šåœ¨ 'B'ï¼‰
    initial = State(monkey='A', box='C', on_box=False, has_banana=False)
    plan = bfs(initial)
    if plan:
        print("æ‰¾åˆ°çš„è¡ŒåŠ¨æ–¹æ¡ˆï¼š")
        step_num = 1
        for action in plan:
            act_name, param = action
            if param is not None:
                print(f"æ­¥éª¤ {step_num}: {act_name} -> {param}")
            else:
                print(f"æ­¥éª¤ {step_num}: {act_name}")
            step_num += 1
    else:
        print("æ²¡æœ‰æ‰¾åˆ°å¯è¡Œçš„è¡ŒåŠ¨æ–¹æ¡ˆã€‚")
````
